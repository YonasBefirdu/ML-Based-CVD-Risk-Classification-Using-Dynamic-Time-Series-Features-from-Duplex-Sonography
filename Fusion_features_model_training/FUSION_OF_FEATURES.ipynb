```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
outer_cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

outer_acc = []
outer_auc = []
outer_f1  = []

inner_cv_results_first_fold = None
for fold, (train_idx, test_idx) in enumerate(outer_cv.split(X, y), start=1):

    print(f"\n========== OUTER FOLD {fold} ==========")

    # -----------------------------
    # Outer split
    # -----------------------------
    X_train = X.iloc[train_idx]
    X_test  = X.iloc[test_idx]
    y_train = y.iloc[train_idx]
    y_test  = y.iloc[test_idx]

    # -----------------------------
    # mRMR on TRAINING DATA ONLY
    # -----------------------------
    selected_features = mrmr_classif(
        X=X_train,
        y=y_train,
        K=15
    )

    X_train_sel = X_train[selected_features]
    X_test_sel  = X_test[selected_features]

    # -----------------------------
    # Inner CV for hyperparameter tuning
    # -----------------------------
    inner_cv = StratifiedKFold(
        n_splits=5,
        shuffle=True,
        random_state=42
    )

    xgb_model = xgb.XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        use_label_encoder=False,
        random_state=42
    )

    grid = GridSearchCV(
        estimator=xgb_model,
        param_grid=param_grid_xgb,
        cv=inner_cv,
        scoring='accuracy',
        n_jobs=-1,
        return_train_score=True
    )

    grid.fit(X_train_sel, y_train)

    if fold == 1:
        inner_cv_results_first_fold = pd.DataFrame(grid.cv_results_)

    best_model = grid.best_estimator_

    # -----------------------------
    # Outer test evaluation
    # -----------------------------
    y_pred = best_model.predict(X_test_sel)
    y_prob = best_model.predict_proba(X_test_sel)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    f1  = f1_score(y_test, y_pred)

    outer_acc.append(acc)
    outer_auc.append(auc)
    outer_f1.append(f1)

    print(
        f"Fold {fold} | "
        f"Accuracy = {acc:.3f}, "
        f"AUC = {auc:.3f}, "
        f"F1 = {f1:.3f}"
    )
outer_acc = np.array(outer_acc)
outer_auc = np.array(outer_auc)
outer_f1  = np.array(outer_f1)

print("\n========== FINAL NESTED CV RESULTS ==========")
print(f"Accuracy : {outer_acc.mean():.3f} ± {outer_acc.std():.3f}")
print(f"AUC      : {outer_auc.mean():.3f} ± {outer_auc.std():.3f}")
print(f"F1-score : {outer_f1.mean():.3f} ± {outer_f1.std():.3f}")
```

```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    f1_score,
    make_scorer
)

# ============================================================
# Hyperparameter grid
# ============================================================
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# ============================================================
# Scoring metrics for INNER CV
# ============================================================
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'auc': 'roc_auc',
    'f1': make_scorer(f1_score)
}

# ============================================================
# Outer CV setup
# ============================================================
outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

outer_acc, outer_auc, outer_f1 = [], [], []
inner_acc, inner_auc, inner_f1 = [], [], []

# ============================================================
# Nested Cross-Validation
# ============================================================
for fold, (train_idx, test_idx) in enumerate(outer_cv.split(X_selected_ts, y), 1):

    # --------------------------
    # Split data
    # --------------------------
    X_train, X_test = X_selected_ts.iloc[train_idx], X_selected_ts.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    # --------------------------
    # Feature subset
    # --------------------------
    X_train_ts = X_train.filter(regex='traditional|shapelet')
    X_test_ts  = X_test.filter(regex='traditional|shapelet')

    # --------------------------
    # mRMR (training only)
    # --------------------------
    selected_features = mrmr_classif(X_train_ts, y_train, K=15)


    X_train_sel = X_train_ts[selected_features]
    X_test_sel  = X_test_ts[selected_features]

    # --------------------------
    # Inner CV (hyperparameters)
    # --------------------------
    inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    rf = RandomForestClassifier(random_state=42)

    grid = GridSearchCV(
        estimator=rf,
        param_grid=param_grid_rf,
        cv=inner_cv,
        scoring=scoring,
        refit='accuracy',
        n_jobs=-1,
        return_train_score=False
    )

    grid.fit(X_train_sel, y_train)

    # --------------------------
    # Extract INNER CV metrics
    # --------------------------
    best_idx = grid.best_index_
    cv_results = grid.cv_results_

    i_acc_mean = cv_results['mean_test_accuracy'][best_idx]
    i_acc_std  = cv_results['std_test_accuracy'][best_idx]

    i_auc_mean = cv_results['mean_test_auc'][best_idx]
    i_auc_std  = cv_results['std_test_auc'][best_idx]

    i_f1_mean  = cv_results['mean_test_f1'][best_idx]
    i_f1_std   = cv_results['std_test_f1'][best_idx]

    inner_acc.append((i_acc_mean, i_acc_std))
    inner_auc.append((i_auc_mean, i_auc_std))
    inner_f1.append((i_f1_mean, i_f1_std))

    # --------------------------
    # Outer test evaluation
    # --------------------------
    best_model = grid.best_estimator_

    y_pred = best_model.predict(X_test_sel)
    y_prob = best_model.predict_proba(X_test_sel)[:, 1]

    o_acc = accuracy_score(y_test, y_pred)
    o_auc = roc_auc_score(y_test, y_prob)
    o_f1  = f1_score(y_test, y_pred)

    outer_acc.append(o_acc)
    outer_auc.append(o_auc)
    outer_f1.append(o_f1)

    print(
        f"Fold {fold} | "
        f"Inner CV → "
        f"Acc={i_acc_mean:.3f}±{i_acc_std:.3f}, "
        f"AUC={i_auc_mean:.3f}±{i_auc_std:.3f}, "
        f"F1={i_f1_mean:.3f}±{i_f1_std:.3f} || "
        f"Outer → "
        f"Acc={o_acc:.3f}, "
        f"AUC={o_auc:.3f}, "
        f"F1={o_f1:.3f}"
    )

# ============================================================
# SUMMARY STATISTICS
# ============================================================
def mean_std(x):
    return np.mean(x), np.std(x)

# Outer (final)
o_acc_m, o_acc_s = mean_std(outer_acc)
o_auc_m, o_auc_s = mean_std(outer_auc)
o_f1_m,  o_f1_s  = mean_std(outer_f1)

# Inner (diagnostic)
i_acc_m, i_acc_s = mean_std([x[0] for x in inner_acc])
i_auc_m, i_auc_s = mean_std([x[0] for x in inner_auc])
i_f1_m,  i_f1_s  = mean_std([x[0] for x in inner_f1])

print("\n================ FINAL RESULTS ================")
print("Outer CV (Generalization Performance)")
print(f"Accuracy : {o_acc_m:.3f} ± {o_acc_s:.3f}")
print(f"AUC      : {o_auc_m:.3f} ± {o_auc_s:.3f}")
print(f"F1-score : {o_f1_m:.3f} ± {o_f1_s:.3f}")

print("\nInner CV (Hyperparameter Tuning Stability)")
print(f"Accuracy : {i_acc_m:.3f} ± {i_acc_s:.3f}")
print(f"AUC      : {i_auc_m:.3f} ± {i_auc_s:.3f}")
print(f"F1-score : {i_f1_m:.3f} ± {i_f1_s:.3f}")

# ============================================================
# Optional: tables for export
# ============================================================
outer_df = pd.DataFrame({
    'Fold': np.arange(1, len(outer_acc) + 1),
    'Accuracy': outer_acc,
    'AUC': outer_auc,
    'F1': outer_f1
})

inner_df = pd.DataFrame({
    'Fold': np.arange(1, len(inner_acc) + 1),
    'Accuracy_mean': [x[0] for x in inner_acc],
    'Accuracy_std':  [x[1] for x in inner_acc],
    'AUC_mean':      [x[0] for x in inner_auc],
    'AUC_std':       [x[1] for x in inner_auc],
    'F1_mean':       [x[0] for x in inner_f1],
    'F1_std':        [x[1] for x in inner_f1],
})

print("\nOuter CV per-fold results:")
print(outer_df)

print("\nInner CV per-fold (best model) results:")
print(inner_df)
```
